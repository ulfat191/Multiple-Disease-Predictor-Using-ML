Train test split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, 
                                                    random_state = 42)
```

Collecting accuracy metrics for classification evaluation

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, plot_confusion_matrix, classification_report, roc_curve, auc
from sklearn.metrics import accuracy_score
from sklearn import metrics
```

Logistic Regression -> model1, fitting data, evaluating performance

```python
from sklearn.linear_model import LogisticRegression
```

```python
# logistic_regression -> model1
model1 = LogisticRegression(random_state=42, max_iter=100)
model1.fit(X_train, y_train)
```

```python
# training score for logistic regression
model1_train_score = model1.score(X_train, y_train)*100
print(f"Training accuracy for Logistic Regression Model = {round(model1_train_score, 2)} %")
```

```python
# predicting outcome for logistic regression model1
model1_predictions = model1.predict(X_test)
```

```python
# evaluating performance for logistic regression model1
model1_scores = (metrics.accuracy_score(y_test, model1_predictions)*100)
print(f"Testing accuracy for Logistic Regression = {round(model1_scores, 1)} %")

# Logistic regression confusion matrix
print("Confusion matrix for Logistic Regression -")
plot_confusion_matrix(model1, X_test, y_test)
plt.show()

# Logistic regression classification report for other evaluation params
print(f"Classification report for Logistic Regression - \n{classification_report(y_test, model1_predictions)}")
```

```python
# roc_curve and auc for logistic regression
model1_fpr, model1_tpr, threshold = roc_curve(y_test, model1_predictions)
model1_auc = auc(model1_fpr, model1_tpr)

plt.figure(figsize= (5, 5), dpi = 200)
plt.plot(model1_fpr, model1_tpr, marker='.', label='Logistic Regression Model (auc = %0.2f)' % model1_auc)
plt.xlabel('FPR')
plt.ylabel('TPR')

plt.legend()

plt.show()
```

SVM -> model6

```python
from sklearn.svm import SVC
```

```python
# setting kernels for SVM
kernel = ['linear', 'poly', 'rbf']

# running loop to train and evaluate
for k in kernel:
    model6 = SVC(kernel=k)
    model6.fit(X_train, y_train)
    print(f"For SVM kernel {k}:")
    model6_predictions = model6.predict(X_test)
    print(f"Training acccuracy = {model6.score(X_train, y_train)*100} %")
    model6_scores = (metrics.accuracy_score(y_test, model6_predictions)*100)
    print(f"Testing accuracy = {round(model6_scores, 2)} %")
    print(f"Confusion matrix -")
    plot_confusion_matrix(model6, X_test, y_test)
    plt.show()
    print(f"Classification report \n {classification_report(y_test, model6_predictions)}")
    
    # roc_curve and auc for SVM    
    model6_fpr, model6_tpr, threshold = roc_curve(y_test, model6_predictions)
    model6_auc = auc(model6_fpr, model6_tpr)

    plt.figure(figsize= (5, 5), dpi = 200)
    plt.plot(model6_fpr, model6_tpr, marker='.', label='SVM Model (auc = %0.2f)' % model6_auc)
    plt.xlabel('FPR')
    plt.ylabel('TPR')

    plt.legend()

    plt.show()  
```

```python
# evaluating performance for SVM Kernels model6
for k in kernel:
    print(f"For kernel = {k}")
    model6_scores = (metrics.accuracy_score(y_test, model6_predictions)*100)
    print(f"Testing accuracy for Naive Bayes  = {round(model5_scores, 1)} %")

    # SVM confusion matrix
    print("Confusion matrix for SVM  -")
    plot_confusion_matrix(model6, X_test, y_test)
    plt.show()

    # SVM report for other evaluation params
    print(f"Classification report for SVM - \n{classification_report(y_test, model6_predictions)}")
```
